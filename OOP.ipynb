```python
class Employee:
    "Common base class for all employees"
    empCount = 0  # Class variable to track number of employees

    def __init__(self, name, salary):  # Constructor method
        self.name = name
        self.salary = salary
        Employee.empCount += 1

    def displayCount(self):
        print("Total Employees:" , Employee.empCount)

    def displayEmployee(self):
        print("Name: %s, Salary: %d" % (self.name, self.salary))


emp1 = Employee("Alice", 50000)#creating instance objects
emp2 = Employee("Bob", 60000)

emp1.displayEmployee()
emp2.displayEmployee()

emp1.displayCount()
```

```python
class Vector:
 def __init__ (self, a, b):
  self.a = a
  self.b = b
 def __str__(self):
    return f'Vector ({self.a}, {self.b})'
 def __add__(self,other):
  return Vector(self.a + other.a, self.b + other.b)
v1 = Vector(2,10)
v2 = Vector(5,-2)
print (v1 + v2)
```

```python
class Employee:
    def __init__(self, name, id):
        self.name = name
        self.id = id

    def updatePhone(self, phone):
        self.phone = phone
        return True

class Developer(Employee):
    def __init__(self, name, id, gitHubLink):      # 3 parameters
        super().__init__(name, id)
        self.gitHubLink = gitHubLink

    def updateGitHubLink(self, link):
        self.gitHubLink = link

class Designer(Employee):
    def __init__(self, name, id, dribbleLink):     # 3 parameters
        super().__init__(name, id)
        self.dribbleLink = dribbleLink

    def updateDribbleLink(self, link):
        self.dribbleLink = dribbleLink


#  Correct — passing 3 arguments each
dev1 = Developer("grace", 7, "github.com/grace")
dev2 = Developer("susan", 9, "github.com/susan")
des1 = Designer("john", 6, "dribbble.com/john")
des2 = Designer("mary", 5, "dribbble.com/mary")

print(dev1.name, dev1.id, dev1.gitHubLink)
print(dev2.name, dev2.id, dev2.gitHubLink)
print(des1.name, des1.id, des1.dribbleLink)
print(des2.name, des2.id, des2.dribbleLink)
```
