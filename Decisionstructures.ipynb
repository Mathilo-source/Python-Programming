```python
#if statements
age=29
if age<18:
  print("Minor")
print(age)
```

```python
age=int(input("Enter your age:"))
if age<18:
  print("Minor")
```

```python
#if...else statements
age=int(input("Enter your age:"))
if age<18:
  print("Minor")
else:
  print("Adult")
```

```python
#elif statements
score=int(input("Enter your score:"))
if score<40 and score>0:
  grade="F"
elif score<50:
  grade="D"
elif score<60:
  grade="C"
elif score<80:
  grade="B"
else:
  grade="A"
print(grade)
```

```python
#nested if statements
marks = int(input("Enter your marks: "))
if marks >= 50:
    if marks >= 70:
        print("Distinction")
    else:
        print("Pass")
else:
    print("Fail")
```

```python
#single statement
num=int(input("Enter a number:"))
if num%3==0:print("Number divisible by 3")
```

```python
#match case
day=int(input("Enter a day:"))
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid day")
```
