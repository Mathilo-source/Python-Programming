```python
#the open() and close() methods
myfile = open("text.txt", "w+")
print ("Name of the file: ", myfile.name)
print ("Closed or not : ", myfile.closed)
print ("Opening mode : ", myfile.mode)
myfile.close()
print ("Closed or not : ", myfile.closed)
```

```python
myfile = open("text.txt", "r+") # Open a file
rstr = myfile.read(5)
print ("Reading up to position 5: \n", rstr)
position = myfile.tell() # Check current position
print ("Current file position : ", position)
myfile.seek(15, 0) # Moving the pointer ahead
rstr = myfile.read(6)
print ("Reading six positions from here : ", rstr) # Close opened file
myfile.close()
```

```python
def countspaces(file):
    fx= open(file)
    count=0
    text=fx.read()
    for w in text:
        if (w==" "): count+=1
    fx.close()
    return count
print(countspaces("text.txt"))
```

```python
def countlove(filepath):
  file=open(filepath,"r")
  text=file.read()
  words=text.split()
  count=words.count("Line")
  print(count)
  return
  file.close()
countlove("text.txt")
```
