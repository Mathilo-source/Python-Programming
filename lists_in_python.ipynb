```python
list6=["orange","mango","tomato","apple","kiwi"]
print(list6)
print(list6[0])#accessing index 0
print(list6[0:4])#index 4 will be excluded
print(list6[1:5])
print(list6[1:])
print(list6[0:3])#itafika tomato
print(list6[:3])#start to tomato
```

```python
list7=[1,2,3,4,5]
print(list7)
print(list7[0])#accessing index 0
print(list7[0:4])#index 4 will be excluded
print(list7[1:5])
print(list7[1:])
print(list7[0:3])#itafika tomato
print(list7[:3])#start to tomato
print(list7[0:5:2])#in an interval of 2
print(list7[-1])
print(list7[-3:-1])
print(list7[::-1])#inareverse
```
