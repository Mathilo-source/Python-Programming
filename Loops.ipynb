```python
#printing a triangle
rows=int(input("Enter the number of rows:"))
for i in range(1,rows+1):
 for j in range(1,i+1):
  print("*",end="")
 print()
```

```python
#printing an upside down triangle
rows=int(input("Enter the number of rows:"))
for i in range(rows,0,-1):#range(start,stop,update)
  for j in range(1,i+1):
    print("*", end="")
  print()
```

```python
#spaced right angled triangle
rows=int(input("Enter the number of rows:"))
for i in range(1,rows+1):
  for j in range(1,rows+1):
    if j<=rows-i:
      print(" ",end="")
    else:
      print("*",end="")
  print()
```

```python
#inverted spaced right angled triangle
rows=int(input("Enter the number of rows:"))
for i in range(1,rows+1):
  for j in range(1,rows+1):
    if j<i:
      print(" ",end="")
    else:
      print("*",end="")
  print()
```

```python
#upside down triangle
rows=int(input("Enter the number of rows:"))
for i in range(1,rows+1):
  for j in range(1,rows+1):
    if j<i:
      print(" ",end="")
    else:
      print("* ",end="")
  print()
```

```python
#triangle
rows=int(input("Enter the number of rows:"))
for i in range(1,rows+1):
  for j in range(1,rows+1):
    if j<=rows-i:
      print(" ",end="")
    else:
      print("* ",end="")
  print()
```

```python
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):

    # Spaces
    for s in range(rows - i):
        print(" ", end="")

    # Stars
    for star in range(2 * i - 1):
        print("*", end="")

    print()
```

```python
rows = int(input("Enter number of rows: "))

# Upper pyramid
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2*i - 1))

# Lower inverted pyramid
for i in range(rows - 1, 0, -1):
    print(" " * (rows - i) + "*" * (2*i - 1))
```

LOOP CONTROL STATEMENTS

```python
#break statement
name = 'university'
for letter in name:
 if letter=='r':
  break
 print(letter)
```

```python
num = 0
while num<10:
 if num==6:
  break
 print(num*num)
 num+=1
```

```python
#continue statement
name = 'university'
for letter in name:
 if letter=='r':
  continue
 print(letter)
```

```python
#pass statement
for letter in 'university':
 if (letter == 'v'):
  pass
 print (letter)
```
