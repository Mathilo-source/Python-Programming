```python
def simple_counter():
    print("Starting...")
    yield 1
    print("Resuming...")
    yield 2
    print("Ending...")

# 1. Initialize: Returns the object, prints nothing
gen = simple_counter()

# 2. First call: Starts and stops at first yield
print(next(gen)) # Output: Starting... 1

# 3. Second call: Resumes and stops at second yield
print(next(gen)) # Output: Resuming... 2
```

```python
def fib(n): # return Fibonacci series up to n
  result = []
  a, b = 0, 1
  while b < n:
    result.append(b)
    a, b = b, a+b
  return result
if __name__== "__main__":
  f=fib(100)
  print(f)
```

```python
def missing_digits(phone_number):
    # Always treat the phone number as a STRING to preserve leading zeros
    phone_str = str(phone_number)

    # Remove any formatting characters (spaces, dashes, +, parentheses)
    phone_str = ''.join(filter(str.isdigit, phone_str))

    # Create a set of digits present in the phone number
    present_digits = set(phone_str)

    # Check digits 0-9 and collect missing ones
    missing = []
    for digit in "0123456789":
        if digit not in present_digits:
            missing.append(int(digit))

    return missing


# ✅ FIX: Pass the phone number as a STRING to preserve leading zero
result = missing_digits("0712345678")
print("Missing digits:", result)

# Works with formatting too
result2 = missing_digits("+254 712-345-678")
print("Missing digits:", result2)
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Male Students', 'Female Students']
sizes = [48, 16]  # Number of students
colors = ['skyblue', 'pink']  # Optional: colors for the slices
explode = (0.1, 0)  # Slightly "explode" the first slice (Male)

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)

# Equal aspect ratio ensures the pie is circular
plt.axis('equal')

# Add title
plt.title("Distribution of Male and Female Students in the Club")

# Show the chart
plt.show()
```
